```python
#EJEMPLO DE LO QUE SE CREA CUANDO EJECUTAS ESTOS COMANDOS INCIALES

# empezamos creando el subplot con el número de gráficas deseadas en su interior
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 5))
fig.suptitle('Relaciones con la Tasa de Natalidad', fontsize=16)
```

```python
# empezamos creando el subplot con el número de gráficas deseadas en su interior
fig, axes = plt.subplots(2, 2, figsize=(15, 10)) 
fig.suptitle("Relaciones con la Tasa de Natalidad", fontsize=16)

# Natalidad vs Mortalidad Infantil
sns.scatterplot(data=df, x='infant_mortality', y='birth_rate', ax=axes[0, 0])
axes[0, 0].set_title("Natalidad vs Mortalidad Infantil")

# Natalidad vs Esperanza de Vida
sns.scatterplot(data=df, x='life_expectancy', y='birth_rate', ax=axes[0, 1])
axes[0, 1].set_title("Natalidad vs Esperanza de Vida")

# Natalidad vs Densidad de Población
sns.scatterplot(data=df, x='population', y='birth_rate', ax=axes[1, 0])
axes[1, 0].set_title("Natalidad vs Densidad de Población")

# Natalidad vs Población Urbana
sns.scatterplot(data=df, x='urban_population', y='birth_rate', ax=axes[1, 1])
axes[1, 1].set_title("Natalidad vs Población Urbana")

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
